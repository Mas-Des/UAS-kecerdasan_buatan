```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Fungsi Keanggotaan Pendapatan
def pendapatan_rendah(x):
    return np.clip((5000000 - x) / 2000000, 0, 1)

def pendapatan_sedang(x):
    return np.clip((x - 3000000) / 2000000, 0, 1)

def pendapatan_tinggi(x):
    return np.clip((x - 5000000) / 2000000, 0, 1)

# 2. Fungsi Keanggotaan Utang
def utang_sedikit(x):
    return np.clip((1000000 - x) / 500000, 0, 1)

def utang_sedang(x):
    return np.clip((x - 1000000) / 1000000, 0, 1)

def utang_banyak(x):
    return np.clip((x - 2000000) / 1000000, 0, 1)

# 3. Fungsi Keanggotaan Riwayat Kredit
def riwayat_buruk(x):
    return np.clip((5 - x) / 2, 0, 1)

def riwayat_cukup(x):
    return np.clip((x - 3) / 2, 0, 1)

def riwayat_baik(x):
    return np.clip((x - 5) / 2, 0, 1)

# 4. Input
pendapatan = 7000000
utang = 3000000
riwayat_kredit = 5

# 5. Plot Fungsi Keanggotaan Pendapatan
x_pendapatan = np.linspace(0, 10000000, 500)
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.plot(x_pendapatan, pendapatan_rendah(x_pendapatan), label="Rendah", color="blue")
plt.plot(x_pendapatan, pendapatan_sedang(x_pendapatan), label="Sedang", color="green")
plt.plot(x_pendapatan, pendapatan_tinggi(x_pendapatan), label="Tinggi", color="red")
plt.axvline(x=pendapatan, color="black", linestyle="--", label=f"Pendapatan = {pendapatan/1000000} Juta")
plt.title("Fungsi Keanggotaan Pendapatan Bulanan")
plt.xlabel("Pendapatan Bulanan (IDR)")
plt.ylabel("Keanggotaan")
plt.legend()

# 6. Plot Fungsi Keanggotaan Utang
x_utang = np.linspace(0, 5000000, 500)
plt.subplot(2, 2, 2)
plt.plot(x_utang, utang_sedikit(x_utang), label="Sedikit", color="blue")
plt.plot(x_utang, utang_sedang(x_utang), label="Sedang", color="green")
plt.plot(x_utang, utang_banyak(x_utang), label="Banyak", color="red")
plt.axvline(x=utang, color="black", linestyle="--", label=f"Utang = {utang/1000000} Juta")
plt.title("Fungsi Keanggotaan Utang")
plt.xlabel("Utang (IDR)")
plt.ylabel("Keanggotaan")
plt.legend()

# 7. Plot Fungsi Keanggotaan Riwayat Kredit
x_riwayat = np.linspace(0, 10, 500)
plt.subplot(2, 2, 3)
plt.plot(x_riwayat, riwayat_buruk(x_riwayat), label="Buruk", color="blue")
plt.plot(x_riwayat, riwayat_cukup(x_riwayat), label="Cukup", color="green")
plt.plot(x_riwayat, riwayat_baik(x_riwayat), label="Baik", color="red")
plt.axvline(x=riwayat_kredit, color="black", linestyle="--", label=f"Riwayat Kredit = {riwayat_kredit}")
plt.title("Fungsi Keanggotaan Riwayat Kredit")
plt.xlabel("Riwayat Kredit")
plt.ylabel("Keanggotaan")
plt.legend()

# 8. Defuzzifikasi: Output Kelayakan Kredit
kelayakan_baik = 0.5
kelayakan_cukup = 0.5
kelayakan_buruk = 0

# Titik representatif dari hasil fuzzy output
baik_titik = 7
cukup_titik = 5
buruk_titik = 3

# Menghitung nilai defuzzifikasi
output_crisp = (kelayakan_baik * baik_titik + kelayakan_cukup * cukup_titik + kelayakan_buruk * buruk_titik) / (kelayakan_baik + kelayakan_cukup + kelayakan_buruk)

# 9. Plot Hasil Defuzzifikasi
plt.subplot(2, 2, 4)
plt.bar(["Baik", "Cukup", "Buruk"], [kelayakan_baik, kelayakan_cukup, kelayakan_buruk], color=['red', 'green', 'blue'])
plt.axhline(y=output_crisp, color='black', linestyle='--', label=f"Defuzzifikasi: {output_crisp:.2f}")
plt.title("Hasil Defuzzifikasi Kelayakan Kredit")
plt.ylabel("Keanggotaan")
plt.legend()

# Tampilkan grafik
plt.tight_layout()
plt.show()
```
